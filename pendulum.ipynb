```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('pendulum_data.csv')
# print(df)
length_in_metre = df['length (cm)'] / 100
period_in_seconds = df['period (seconds)']
uncertainty_in_period = df['uncertainty in period (seconds)']
```

### Plot graph before fittign the line

```python
plt.scatter(length_in_metre, period_in_seconds)
```

### Find g

```python
def f(l,g):
	return (2*np.pi) * np.sqrt(l/g)

popt, pcov = curve_fit(f, length_in_metre, period_in_seconds, sigma=uncertainty_in_period)
g = popt[0]
uncertainty_g = np.sqrt(pcov[0][0])
g, uncertainty_g
```

So, Answer of g = $10.053894877764414 \pm 0.09580450073848856$

### Plot fit line in graph length (metre) Vs. period (second)

```python
plt.plot(length_in_metre, f(length_in_metre,g))
plt.errorbar(length_in_metre, period_in_seconds, yerr=uncertainty_in_period, fmt='.', color="red")
plt.xlabel("Length (metre)")
plt.ylabel("Period (second)")
plt.show()
```
